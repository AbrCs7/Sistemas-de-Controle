```python
import control as ctl
import matplotlib.pyplot as plt
import numpy as np
```

criação da FT

```python
P_s = ctl.tf([1.,2.], [1.,1., -5])
R_s = ctl.tf([1.], [1.,0.])
C_s = R_s * P_s;
print(' FT = ', C_s)
```

```python
ps_ma = ctl.poles(C_s)
zs_ma = ctl.zeros(C_s)
print('polos = ', ps_ma)
print('zeros = ', zs_ma)
```

Fazer doagrama pz

```python
plt.figure()
PZ = ctl.pzmap(C_s, 1, 1, 'diagrama pz de C_s')
```

Resposta ao degrau

```python
Tsim = 10;
quais_T = np.linspace(-1., Tsim, 1000)
T_mf, yout_mf = ctl.step_response(P_s, quais_T[quais_T > 0])
```

Calcula um degrau unitário

```python
degrau = np.ones_like(quais_T)
degrau[quais_T < 0] = 0;
```

Plota os resultados

```python
plt.figure()
plt.plot(T_mf, yout_mf, 'k-')
plt.plot(quais_T, degrau, 'r-')
plt.xlabel('tempos (s)')
plt.legend(['resposta ao degrau','degrau unitário'])
plt.grid(); plt.title('Sistema de 2ª ordem');
```

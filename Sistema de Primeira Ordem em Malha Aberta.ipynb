Importar as bibliotecas (toolbox necessárias:

```python
import control as ctl
import matplotlib.pyplot as plt
import numpy as np
```

Vamos criar as variáveis R, C e Tsim e atribuir valores a elas

```python
R = 1.; #Resistência
C = 1.; #Capacitância
Tsim = 10; #Tempo de simulação
```

Vamos criar os vetores que serão os polinômios: "1" e "RCs + 1"


```python
numerador = [1.]
denominador = [R*C,1.]
```

A partir desses vetores vamos criar a função de transferência

```python
H = ctl.tf(numerador, denominador)
print('FT de malha aberta = ', H) #Mostra a função de transferência
```

Vamos agora calcular a resposta ao degral (step) para Tsim e H definidos aciama

```python
T,yout = ctl.step_response(H, Tsim)
```

Plotar a resposta ao degrau

```python
plt.plot(T,yout,'b-')
```

Calcula e plota um degral unitário, de -1 até 10 para comparar com a saída

```python
T2 = np.linspace(-1.,10.,1000)
degrau = np.ones_like(T2)
degrau[T2<0]=0;
plt.plot(T2, degrau, 'r-')
```

Plortar titulos nos eixos, uma legenda e um frid para deixar o gráfico mais apresentável

```python
plt.ylabel('V (V)')
plt.xlabel('tempo (s)')
plt.legend(['resposta ao degrau','degrau unitário'])
plt.grid()
```
